```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
import numpy as np

X_vif = x_train_scaled  # use scaled features

vif_data = pd.DataFrame()
vif_data["Feature"] = X_train.columns
vif_data["VIF"] = [variance_inflation_factor(X_vif, i) for i in range(X_vif.shape[1])]

vif_data
```

Variance Inflation Factor (VIF) analysis showed that all predictors had VIF values below 2, indicating no evidence of multicollinearity. This confirms that Emergency_Rate, Waited_12hr_Rate, and Log_Total_Attendances contribute independent information to the model and that coefficient estimates from linear regression are stable and reliable.

```python
#mixed model
import statsmodels.formula.api as smf
```

```python
model_mixed = smf.mixedlm(
    "Over4hr_Percentage ~ Emergency_Rate + Waited_12hr_Rate + Log_Total_Attendances",
    df_new,
    groups=df_new["Org name"]
)

result_mixed = model_mixed.fit()

print(result_mixed.summary())
```

The mixed-effects model reveals significant between-trust heterogeneity (variance: 154.5), confirming that baseline congestion varies by location. While 12-hour waits and attendance volume remain positive predictors of breach rates, the Emergency_Rate coefficient becomes negative once trust-level effects are controlled. This shift suggests that within-trust fluctuations behave differently than cross-sectional comparisons, emphasizing the need to distinguish structural heterogeneity from temporal variation in healthcare modeling.

```python
# Mixed Effects
import statsmodels.formula.api as smf

model_ols = smf.ols(
    "Over4hr_Percentage ~ Emergency_Rate + Waited_12hr_Rate + Log_Total_Attendances",
    data=df_new
).fit()

print(model_ols.summary())
```

```python
print("OLS AIC:", model_ols.aic)
print("OLS Log-Likelihood:", model_ols.llf)

print("Mixed AIC:", result_mixed.aic)
print("Mixed Log-Likelihood:", result_mixed.llf)
```

```python
model_mixed_ml = smf.mixedlm(
    "Over4hr_Percentage ~ Emergency_Rate + Waited_12hr_Rate + Log_Total_Attendances",
    df_new,
    groups=df_new["Org name"]
).fit(reml=False)

print(model_mixed_ml.summary())
```

The mixed-effects model fits dramatically better than the standard OLS regression. The large improvement in log-likelihood and substantial reduction in AIC indicate that accounting for trust-level random intercepts significantly improves model fit. This confirms that observations are clustered within trusts and that baseline congestion varies meaningfully across institutions. Ignoring this hierarchical structure leads to substantial model misspecification.

```python
model_mixed_interaction = smf.mixedlm(
    "Over4hr_Percentage ~ Emergency_Rate * Log_Total_Attendances + Waited_12hr_Rate",
    df_new,
    groups=df_new["Org name"]
).fit(reml=False)

print(model_mixed_interaction.summary())
```
